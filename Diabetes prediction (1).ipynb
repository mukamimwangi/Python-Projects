```python
#Correlation with target column(outcome)
df.corr()['Outcome'].sort_values()
```

```python
#Separate target column from the entire data
X = df.drop(columns= 'Outcome', axis=1)
Y = df['Outcome']
```

```python
#Histogram for all numerical entries in the data
X.hist(figsize=(10,8), bins=20)
plt.tight_layout()
plt.show()
```

```python
#Standardize the data, load the Scaler function
scaler = StandardScaler()
```

```python
#Fit the data in the scaler
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python

#Load the data
X = standardized_data
Y = df['Outcome']
```

```python
#Perform train/test/split on the data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
#Print the shape of the train&test data
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#1st, train the data on a logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
logreg_pred = logreg.predict(X_test)
logreg_acc = accuracy_score(logreg_pred, Y_test)
print("Test Accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
#Print the classification report
print(classification_report(Y_test, logreg_pred))
```

```python
# 2nd, train the data on a Support Vector Machine (svm) model
svmModel = svm.SVC(kernel='linear')
svmModel.fit(X_train, Y_train)
svmModel_pred = svmModel.predict(X_test)
svmModel_acc = accuracy_score(svmModel_pred, Y_test)
print("Test Accuracy: {:.2f}%".format(svmModel_acc*100))
```

```python
print(classification_report(Y_test, svmModel_pred))
```

```python
#Make a modelprediction using svmModel
input_data = (1,166,72,19,175,25.8,0.587,51)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
scalar_data = scaler.transform(input_data_reshaped)
```
