```python
import pandas as pd #to read csv files
import numpy as np  #to arrange data on x axis
import matplotlib.pyplot as plt #to plot the data
plt.style.use('ggplot')
%matplotlib inline
from matplotlib import style #add style to the plot
```

```python
time_data={'facebook':120, 'youtube':90,'instagram':40, 'snapchat':28}
```

```python
social_media=list(time_data.keys()) #x-axix
```

```python
time_spent=list(time_data.values())#y-axis
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
color=['purple','blue','red','green']
plt.bar(social_media,time_spent, color=color,width=0.5)
plt.xlabel('Social media')
plt.ylabel('Time spent')
plt.title('Average time spent on Social media')
```

```python
fb=(120,90,110,100,95)
yt=(80,90,75,85,95)
ig=(50,40,55,35,45)
sc=(20,30,35,25,85)
```

```python
xpos=np.arange(len(fb))
xpos
```

```python
style.use('ggplot')
plt.figure(figsize=(10,7))
barWidth=0.2
plt.bar(xpos+0.2,fb,color='red',width=barWidth,label='fb')  
plt.bar(xpos+0.4,yt,color='blue',width=barWidth,label='yt') 
plt.bar(xpos+0.6,ig,color='green',width=barWidth,label='ig')
plt.bar(xpos+0.8,sc,color='orange',width=barWidth,label='sc')
plt.xlabel('Day')
plt.ylabel('Time spent in minutes')
plt.legend()
plt.xticks(xpos+0.3,('Monday','Tuesday','Wednesday','Thursday','Friday'))
plt.title('Time spent on social media')
```
